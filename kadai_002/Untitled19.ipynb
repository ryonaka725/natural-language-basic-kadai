```python
import re

email = 'abc@xxx.com'
match_string = re.match(r'[a-z]+@[a-z]+\.[a-z]+', email)
print(match_string)

print(match_string.start())
print(match_string.end())
print(match_string.span())
print(match_string.group())
```
